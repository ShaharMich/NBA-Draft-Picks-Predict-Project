```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import statistics
```

```python
from sklearn import linear_model
from sklearn import metrics,preprocessing
from scipy.special import expit
from scipy.special import logit
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
Draft_Picks_df = pd.read_csv("All_Draft_Picks_Update.csv")
Draft_Picks_df = Draft_Picks_df.drop("Unnamed: 0",axis = 1)
Draft_Picks_df = Draft_Picks_df.reset_index()
Draft_Picks_df = Draft_Picks_df.drop("index",axis = 1)
```

```python
#we took the most important indices for each player stats, and put in a new dataframe
new_df = Draft_Picks_df[["PTS","FGP","TPP","AST","STL","BLK","TRB","TOV","Height","Weight","Nationality","NIG","MIN","GP","Age","FTP"]]
columnsToEncode = list(new_df.select_dtypes(include=['category', 'object']))
le = LabelEncoder()
for feature in columnsToEncode:
    try:
        new_df[feature] = le.fit_transform(new_df[feature])
    except:
        print('Error encoding ' + feature)
print(new_df)
```

```python
preclist = list()
count0=0
count1 = 0
#We trained and tested our machine to check how accurate we are
for i in range(100):
    Xtrain,Xtest,ytrain,ytest = train_test_split(new_df,Draft_Picks_df["is_Drafted"])
    lrm=linear_model.LogisticRegression()
    lrm.fit(Xtrain,ytrain)
    y_pred = lrm.predict(Xtest)
    metrics.confusion_matrix(ytest, y_pred)
    print("precision is:",metrics.precision_score(ytest, y_pred))
    preclist.append(metrics.precision_score(ytest, y_pred))
print(statistics.mean(preclist))
```
