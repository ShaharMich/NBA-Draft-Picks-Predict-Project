```python
import numpy as np
import os
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
# We loaded the csv file that we created in the crawling section
Draft_Picks_df = pd.read_csv('All_Draft_Picks.csv')
Draft_Picks_df = Draft_Picks_df.drop("Unnamed: 0",axis = 1)
print(Draft_Picks_df)
```

```python
for i in range(Draft_Picks_df.shape[0]):
    for j in range(Draft_Picks_df.shape[1]):
        if Draft_Picks_df.iloc[i,j] == 'N/A':
            Draft_Picks_df[i,j] = np.nan
```

```python
# We dropped the players that are missing data
Draft_Picks_df = Draft_Picks_df.dropna(axis = 0, how = 'any')
print(Draft_Picks_df)
```

```python
Draft_Picks_df = Draft_Picks_df.drop_duplicates()
print(Draft_Picks_df)
```

```python
# We created new list, if player got drafted, we put in his cell 1, if he did not drafted, we put 0
is_Drafted = list()
for i in Draft_Picks_df['Pick']:
    if i == 0:
        is_Drafted.append(0)
    else:
        is_Drafted.append(1)
        
Draft_Picks_df['is_Drafted'] = is_Drafted
```

```python
# We calculate some of the attack stats into ATK list, defensive stats into DEF list, and phisycal stats into BMI list
BMI = list()
ATK = list()
DEF = list()
for i in range (Draft_Picks_df.shape[0]):
    BMI.append(Draft_Picks_df.iloc[i,8]/((Draft_Picks_df.iloc[i,7]/100)**2))
    DEF.append((Draft_Picks_df.iloc[i,26]+Draft_Picks_df.iloc[i,25]+Draft_Picks_df.iloc[i,22])/3)
    if(Draft_Picks_df.iloc[i,27] == 0):
        ATK.append(((Draft_Picks_df.iloc[i,11]+Draft_Picks_df.iloc[i,21])+Draft_Picks_df.iloc[i,24]+Draft_Picks_df.iloc[i,14])/4)
    else:    
        ATK.append(((Draft_Picks_df.iloc[i,11]+Draft_Picks_df.iloc[i,21])+(Draft_Picks_df.iloc[i,24]/Draft_Picks_df.iloc[i,27])+Draft_Picks_df.iloc[i,14])/4)



Draft_Picks_df['BMI'] = BMI
Draft_Picks_df['ATK'] = ATK
Draft_Picks_df['DEF'] = DEF
Draft_Picks_df
```

```python
Draft_Picks_df.to_csv("All_Draft_Picks_Update.csv")
Draft_Picks_df.to_excel("All_Draft_Picks_Update.xlsx")
```
